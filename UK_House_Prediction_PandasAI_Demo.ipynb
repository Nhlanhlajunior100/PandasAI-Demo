## Data Cleaning

#### Requirements
* Drop Unnamed column
* Rename columns with units where necessary
* Remove text from Price paid, Floor area, euro per square metre (float)
* Separate the house type and the bedroom count
* Do some encoding to location and type columns. Binary & One-Hot respectively
* Drop the property & date column (our analysis is on city level). This is to reduce the high dimensionality of the property feature.
* Drop last sold date colum (subjective factor that we'll consider for improving the model)
* Consider the best way to handle nulls for model accuracy (The dataset has )
* Check for outliers. Extreme values can skew our model. Statistical methods include finding the percentage of values that deviate from the mean by 1 or more standard deviations. 



```python
uk_houses.head()
```

```python
#dropping 'Unnamed: 0' col (if it exists)

try:
    uk_houses= uk_houses.drop(['Unnamed: 0'], axis=1)
except KeyError:
    print("Column's been deleted")

#uk_houses.head()
```

```python
uk_houses.rename(columns = {'Price paid': 'price_paid_pounds' ,\
                            'Last sold': 'last_sold',\
                                  'Floor area':'floor_area_sqm',\
                                  uk_houses.columns[4]:'pound_per_sqm'},\
                        inplace=True)
uk_houses.head()
```

```python
# Removing text from first 3 columns
uk_houses[['price_paid_pounds', 'floor_area_sqm', 'pound_per_sqm']] = uk_houses[['price_paid_pounds', 'floor_area_sqm', 'pound_per_sqm']].apply(lambda x: x.str.replace('\D', '', regex=True))
uk_houses.head()
```

```python
# splitting the bed count column into house type and bed count
uk_houses[['house_type', 'bed_count']] = uk_houses['Type'].str.split('(', n=1, expand=True)


# Removing text from bed count
uk_houses['bed_count'] = uk_houses['bed_count'].str.replace('\D', '', regex=True)

#dropping original type column
uk_houses = uk_houses.drop(['Type'], axis=1)
uk_houses.head()
```

```python
uk_houses.dtypes
```

```python

uk_houses.isnull().sum()
```

floor area and bed count are the main causes of null results with 2246 & 3420 null values - the pound per sqm is based on the floor area. In this case we have a few options, either replace the null values or frop them entirely. Since this dataset is quite large we'll drop these values in rows where both the the floor area and bedcount are null, then based on the floor area we replace with the modal bed count and replace the floor area with the average for that bed count.

```python
uk_houses_sub = uk_houses.dropna()
```

```python
uk_houses_sub.isnull().sum()
```

```python
uk_houses_sub.dtypes
```

```python
uk_houses_sub.bed_count.unique()
```

```python
uk_houses_sub.isin(['']).sum()
```

```python
uk_houses_sub.drop(index= uk_houses_sub[uk_houses_sub['bed_count']==''].index, inplace=True)
```

```python
uk_houses_sub = uk_houses_sub.astype({
                            'last_sold': 'string',\
                            'property':'string',\
                            'price_paid_pounds': 'float64',\
                            'floor_area_sqm': 'float64',\
                            'pound_per_sqm': 'float64',\
                            'location': 'string',\
                            'house_type': 'string',\
                            'bed_count':'int8'})
```

## Model Training
#### Feature Selection
Since our analysis is on city level we can drop the property column when training the model.This is to reduce the high dimensionality of the property feature.  We'll use the more granular locations when improving the model. The last sold date is a more granular and subjective factor, for this project we're also dropping it.

```python
try:
    uk_houses_sub = uk_houses_sub.drop(['property', 'last_sold'], axis=1)
    print("Done!")
except KeyError:
    print("The columns have been deleted")

uk_houses_sub.head()
```

#### Feature Encoding
For our purposes we'll use One-Hot-Encoding for the property type and binary encoding for the location. When eveluating the model performance we'll consider target encoding for the location. 

```python
#One-Hot-Encoding
#Convert house_type column into numeric using one hot encoding (dummies) then append to uk_houses_sub
dummies = pd.get_dummies(uk_houses_sub['house_type'], prefix='type' ).astype(int)

uk_houses_encoded = pd.concat([uk_houses_sub.drop(['house_type'], axis=1), dummies], axis='columns')
uk_houses_encoded.head()
```

```python
#pip install category_encoders
```

```python
#Binary Encoding
#Convert location column into numeric using binary encoding

#first we instantiate a binary encoder 
import category_encoders as ce

binary_encoder = ce.binary.BinaryEncoder('location')
binary_encoder
```

```python
#now we fit & transform
binary_encoded = binary_encoder.fit_transform(uk_houses_encoded['location'])
binary_encoded.head()
```

```python
#Concat binary encoded and one hot encoded dataframes to uk_houses_encoded
uk_houses_encoded = pd.concat([uk_houses_encoded.drop(['location'], axis=1), binary_encoded], axis='columns')
uk_houses_encoded.head()
```

```python
uk_houses_encoded.dtypes
```

#### Model Fitting
First we need to identify out target and features. Our target is the variable that we want to predict. This is the price paid. All other columns in uk_houses_encoded will be our features. 

```python
X = uk_houses_encoded.drop('price_paid_pounds',axis='columns')
y = uk_houses_encoded.price_paid_pounds
```

```python
#Train-test split (80%-20%)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X , y ,test_size = 0.2, random_state=10)
```

```python
X_train.dtypes
```

```python
X_train.describe()
```

```python
X_test.describe()
```

### Multiple Linear Regression
Assuming a linear relationship between the input features and the target variable, we will fit a linear regression model from sklearn.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
m_score = model.score(X_test , y_test)
```

```python
summary = f"we get a score of{m_score: .4f} which signifies that our Linear Regression model can explain \
approximately{m_score*100: .2f}% of the variance in real estate prices."

print(summary)
```
